```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').value_counts(normalize = True)
```

```python
results = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").value_counts(normalize = True))
```

```python
py_percent_df = pd.DataFrame(results)
py_percent_df.rename(columns = {"Country" : "Country", "LanguageWorkedWith" : "KnowsPython"}, inplace = True)
py_percent_df
```

```python
tot_lang = country_grp['LanguageWorkedWith'].count()
tot_lang
```

```python
knows_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").sum())
knows_python
```

```python
python_df = pd.concat([tot_lang,knows_python], axis = "columns", sort = False)
python_df.columns = ['Total', 'KnowsPython']
python_df['KnowsPython(%)'] = python_df['KnowsPython'] / python_df['Total'] * 100
python_df
```

```python
python_df.drop(columns = ['Total', 'KnowsPython'], inplace = True)
python_df[['KnowsPython(%)']].round(2)
```
